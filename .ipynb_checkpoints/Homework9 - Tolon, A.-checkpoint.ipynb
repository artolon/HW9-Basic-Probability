### 1.	Why is assuming normal data an important consideration when answering questions about probability?

Assumptions of normality allow us to complete different probabiltiy calculations with our data that we would not be able to do otherwise. We rely on these assumptions in order to calculate probabilities of observing a particular mean and/or difference between means. We further rely on these assumptions when producing statistical models and other tests. However, there is definitely danger in assuming normality, such as when trying to generalize findings to the whole populations and/or when using a statistical model (that requires assumptions of normality) to predict something that is not actually normally distributed.

### 4.	What is the probability that the price for an Airbnb is under $50?

```python
# Calculate the probability of a price being below $50
price_under50 = norm.cdf(50, mean, std)
price_under50
    # Probability is approximately 29.28%
```

### Explain The Monty Hall problem in your notebook in markdown in under 250 words.

The Monty Hall problem is a paradoxical statistical scenario. 

The scenario, based off a game show, involves three doors. A prize is behind one of the doors. The other two doors have a goat. A contestant is asked to choose a door at random (hoping for the prize). The host then reveals one of the other two doors to the contestant. The host always reveals a door with a goat. After this reveal, the host asks the contestant if they would like to switch doors. The statistical question is, "Should the contestant switch doors?" The problem is a paradox because one might assume that the probability of getting the prize is always 1/3, regardless of whether they switch or not. However, the mathematical probability of them winning IF they switch after asked is 2/3, making it the wiser choice. 

Here is why it is advantageous to switch:

Before any doors are revealed, the probability of initially selecting the correct door (i.e. winning) is 1/3. If you always stick to your selection (don't switch), the probability will remain 1/3 that you win. However, this also means you have a 2/3 chance of initially picking the wrong door (losing). If you initially pick wrong (which again, is a 2/3 chance) you will ALWAYS win if you switch. This is because the host reveals one of the wrong doors, which means the only remaining door MUST be correct. Because you have a 2/3 chance of initially picking the wrong door AND because you will always win if you switch in that scenario, it is advantageous to switch your selection.
